
## Exercise 4 - Six Boxes Toy Model : inference

• The six boxes toy model

• Labeling the boxes as follows:

![six boxes](six_boxes.png)

• write a program in R that:

1) allows the user to insert the color of a randomly extracted box and

3) prints on the standard output the probability of selecting each box

4) plots the probability for each box as a function of the extraction step


```r
library(ggplot2)
library(gridExtra)
library(repr)

options(repr.plot.width=50, repr.plot.height=25)

nboxes <- 6

j <- 0:(nboxes - 1)

in_prob <- rep(1/nboxes , nboxes)

prob <- data.frame(in_prob)
prob <- data.frame(t(prob))
colnames(prob) <- c('H_0', 'H_1', 'H_2', 'H_3', 'H_4', 'H_5')
rownames(prob) <- 1



repeat{
  
color <- readline(prompt = "\n Insert the color of the extracted stone  [b for black , w for white]. \n Otherwise, if the extraction phase is finished, insert anything else :) \n ")

if(toupper(color == 'w')){
  pr <- j/max(j)
}
else if (toupper(color == 'b')){
  pr <- (max(j) - j) / max(j)
}

else{
  cat('\n Extraction phase terminated \n')
  break
}
  
probability <- (pr * prob[nrow(prob), ])

norm_prob <- probability/(sum(probability))

prob[nrow(prob)+1, 1:nboxes] <- norm_prob
    
    
}

x_ax <- 0:(nrow(prob)-1)

grid.arrange(plot_func(1), plot_func(2), plot_func(3),plot_func(4), plot_func(5), plot_func(6), ncol=3)


cat('\n Probabilities during each extraction: ')
prob
```

## Exercise 5 - Six Boxes Toy Model : simulation

• consider again the six boxes toy model of the previous exercise and write a simulation
program that:

1) selects a random box

2) makes random sampling from the box

3) prints on the standard output the probability of selecting each box

4) plots the probability for each box as a function of the number of trial


```r
library(ggplot2)
library(gridExtra)
library(repr)


options(repr.plot.width=50, repr.plot.height=25)

plot_func <- function(column, df=prob){
   plot <- ggplot() + theme_linedraw() + geom_point(aes(x=x_ax, y=df[,column]), size=12 , color='black', pch=21, fill='steelblue4') +
    labs(title= sprintf('Probability of the box %s', colnames(df)[column]), x='Step', y='Probability') + 
     theme(plot.title = element_text(size=60), 
                  axis.text.x = element_text(size=40), 
                  axis.text.y = element_text(size=40),
                  axis.title.x = element_text(size=52),
                  axis.title.y = element_text(size=52))
}




nboxes <- 6
boxes <- 0:5
j <- 0:5


#initial box probabilities
in_prob <- rep(1/nboxes , nboxes)


prob <- data.frame(in_prob)
prob <- data.frame(t(prob))
rownames(prob) <- 1

prob[1,ncol(prob)+1] <- '-'

colnames(prob) <- c('H_0', 'H_1', 'H_2', 'H_3', 'H_4', 'H_5','Extracted_Stone')




n_draws <- readline(prompt="How many extractions want to be done? ")

#sampling the box
random_box <- sample(boxes, size = 1)


for(extrac in seq(1,n_draws,1)){
    
    ex_stone <- sample(boxes, size = 1)    #Extract a number between [0,5]
                                            #if the number is less than random_box, the stone is white 
                                                # [0,5] corresponds to the number of white stone of the boxes
    if (ex_stone < random_box){
     pr <- (j/max(j))
     col <- 'WHITE'
    } 
    
    else if (ex_stone >= random_box){                #else, if the number extracted is >= the stone is black
     pr <- (max(j) - j) / max(j)
     col <- 'BLACK'
}
 
probability <- (pr * prob[nrow(prob), 1:nboxes ])

norm_prob <- probability/(sum(probability))

prob[nrow(prob)+1,] <- c(norm_prob, col)
    
    
}

x_ax <- 0:(nrow(prob)-1)

grid.arrange(plot_func(1), plot_func(2), plot_func(3), plot_func(4), plot_func(5), plot_func(6), ncol=3 )

cat('\n Probabilities during each extraction: ')
prob

```
